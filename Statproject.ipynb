```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import random
```

```python
#Choose 500 nomber between 0 to 2 randomly
```

```python
sample_size = 500
numbers =np.random.uniform(0,2,500)
print(f"random numbers are :",numbers)
```

```python
#Calculate its average 
```

```python
sample_mean = np.mean(numbers)
print(f"Average of the 500 numbers: {sample_mean: .4f}")
variance = np.var(numbers)
print(variance)
```

```python
#Draw graph of distribution of this 500 numbers
```

```python
plt.figure(figsize=(8,3))
plt.hist(numbers,bins=10,color="lightblue",edgecolor="black")
plt.title("Distribution of 500 random numbers(0-2)")
plt.xlabel("Value")
plt.ylabel("Density")
```

```python
#Demonstrate Central Limit Theorem and Normalization
```

```python
#Generate many samples mean to abserve the sample distribution
```

```python
num_samples = 1000
samples_size = 500
means = []
for _ in range(num_samples):
    samples = np.random.rand(num_samples,samples_size)
    means.append(np.mean(samples))
print(means)
```

```python
#Calculate the mean and standard deviation of these sample means
```

```python
mu_means=np.mean(means)
sigma_means = np.std(means)
variance = np.var(means)
print(f"Mean of the sample means:{mu_means:.4f}")
print(f"Standard devaiation of the sample means:{sigma_means:.4f}")
print(variance)
```

```python
#Normalise the sample means
```

```python
print("Z = (X - mu)/sigma")
```

```python
normalized_means = (np.array(means) - mu_means)/sigma_means
```

```python
#Draw a graph of distribution of sample mean and the normalisation distribution
```

```python
plt.hist(normalized_means,bins = 40,density = True ,color = "lightblue",edgecolor = "black",label = "Normalized Sample Means")

#Define x values for the normal distribution curve
x = np.linspace(-6,10,500)

p = norm.pdf(x,0,1)
plt.plot(x,p,'k',linewidth = 3,label = 'Standard Normal Distribution')

plt.title("Normalized Distribution of Sample Mean(CLT)")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.legend()

plt.show()
```

```python
print(f"\nMean of sample means :{mu_means:.4f}")
print(f"Standard deviation of sample means (Standard Error):{sigma_means:.4f}")
print(f"Variance of sample means:{variance:.9f}")
```
